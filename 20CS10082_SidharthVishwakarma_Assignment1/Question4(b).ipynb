Importing the nesessary libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

Import the image and convert it to a numpy array

```python
image = cv.imread('image0.jpg', 0)
# image=np.array(image)
```

Function to calculate the convolution of an Image with a kernel

```python
def convolution(image, kernel):
    # Get image dimensions
    m = image.shape[0]
    n = image.shape[1]
    # Get kernel dimensions
    p = kernel.shape[0]
    q = kernel.shape[1]
    # Create a new image to store the convolution
    new_image = np.zeros((m+p-1, n+q-1))
    # Loop through the image
    for i in range(m):
        for j in range(n):
            # Loop through the kernel
            for k in range(p):
                for l in range(q):
                    # Calculate the convolution
                    new_image[i+k][j+l] += image[i][j]*kernel[k][l]
    return new_image
```

Plot the Original Image

```python
# Display the original image
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.show()
```

Calculate the convolution of the image with the kernel

```python
convoluted_image = convolution(image, np.array([[1, -1]]))
# Display the convoluted image
plt.imshow(convoluted_image, cmap='gray')
plt.title('Convoluted Image')
plt.show()
```

Repeating the same for 1 and 9 image data from MNIST dataset

```python
image_1 = cv.imread('image1.jpg', 0)
convoluted_image_1 = convolution(image_1, np.array([[1, -1]]))
image_9 = cv.imread('image9.jpg', 0)
convoluted_image_9 = convolution(image_9, np.array([[1, -1]]))
plt.subplot(2, 2, 1)
plt.imshow(image_1, cmap='gray')
plt.title('Original Image')
plt.subplot(2, 2, 2)
plt.imshow(convoluted_image_1, cmap='gray')
plt.title('Convoluted Image')
plt.subplot(2, 2, 3)
plt.imshow(image_9, cmap='gray')
plt.subplot(2, 2, 4)
plt.imshow(convoluted_image_9, cmap='gray')
plt.show()
```
